# Logistic Regression Model

## Imports

```python
import pandas as pd
import numpy as np

from sklearn import linear_model as lm
from sklearn import metrics as mt
```

## Dataset

```python
# Dados Treinamento
X_train = pd.read_csv("/home/ds_deivisson/repos/datasets/X_training_classificacao.csv")
y_train = pd.read_csv("/home/ds_deivisson/repos/datasets/y_training_classificacao.csv")

# Dados Validacao
X_val = pd.read_csv("/home/ds_deivisson/repos/datasets/X_validation_classificacao.csv")
y_val = pd.read_csv("/home/ds_deivisson/repos/datasets/y_validation_classificacao.csv")

# Dados Teste
X_test = pd.read_csv("/home/ds_deivisson/repos/datasets/X_test_classificacao.csv")
y_test = pd.read_csv("/home/ds_deivisson/repos/datasets/y_test_classificacao.csv")
```

## Features/ Label

```python
y_train_class = y_train.iloc[:, 0].to_numpy()
y_val_class = y_val.iloc[:, 0].to_numpy()
y_test_class = y_test.iloc[:, 0].to_numpy()
```

## Logistic regression model - Dados Treino

```python
# define
model = lm.LogisticRegression(max_iter=10000)

# training
model.fit(X_train, y_train_class)

# performance
yhat_train = model.predict(X_train)
f1 = mt.f1_score( y_train, yhat_train)

print ('Acuracia: {}'.format(mt.accuracy_score( y_train, yhat_train)))
print ('Precisao: {}'.format(mt.precision_score( y_train, yhat_train)))
print ('Recall: {}'.format(mt.recall_score( y_train, yhat_train)))
print ('F1-Score: {}'.format(mt.f1_score( y_train, yhat_train)))
       
```

## Logistic regression model - Dados Validação

```python
# define
model = lm.LogisticRegression(max_iter=10000)

# training
model.fit(X_train, y_train_class)

# performance
yhat_val = model.predict(X_val)

print ('Acuracia: {}'.format(mt.accuracy_score( y_val_class, yhat_val)))
print ('Precisao: {}'.format(mt.precision_score( y_val_class, yhat_val)))
print ('Recall: {}'.format(mt.recall_score( y_val_class, yhat_val)))
print ('F1-Score: {}'.format(mt.f1_score( y_val_class, yhat_val)))
       
```

## Logistic regression model - Dados Teste

```python
# concatenação dos dados de treino e validação
X_final_train = pd.concat([X_train, X_val])
y_final_train = np.concatenate((y_train_class, y_val_class))
```

```python
# define
model = lm.LogisticRegression(max_iter=10000)

# training
model.fit(X_final_train, y_final_train)

# performance
yhat_final = model.predict(X_test)

print ('Acuracia: {}'.format(mt.accuracy_score( y_test, yhat_final)))
print ('Precisao: {}'.format(mt.precision_score( y_test, yhat_final)))
print ('Recall: {}'.format(mt.recall_score( y_test, yhat_final)))
print ('F1-Score: {}'.format(mt.f1_score( y_test, yhat_final)))
```
